### Подготовка датафрейма

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
size = 8737

df = pd.DataFrame(
    {
        'time': [*pd.date_range('2021-01-01', '2021-12-31', freq='1h')][:size],
        'company_id': np.random.choice(1000, size),
        'category': np.random.choice(20, size),
        'amount': np.random.choice(3, size),
        'tariff': np.random.choice(3, size),
        'validity': np.random.choice(4, size),
    },
    index=[*range(size)],
)
```

```python
df['category'] = df.category.astype('category')
new_categorits = {
    0: 'Service',
    1: 'Hardware/Garden/Home Improvement',
    2: 'Office Supploes',
    3: 'Pets Care',
    4: 'Home/Furniture',
    5: 'Streaming/Online classes/Software license keys',
    6: 'Apparle/Shoes/Accessories',
    7: 'Automovtive',
    8: 'Art/Photography',
    9: 'Health/Beauty',
    10: 'Tobacco/Vape/Smokinh/Marijuana releated products',
    11: 'Adult',
    12: 'Gifts/Flowers/Collctibles',
    13: 'Electronics/Computers',
    14: 'Toys/Crafts/Hobbies',
    15: 'Sports/Fitness/Outdoors',
    16: 'Jewelry',
    17: 'Subscription-based products of servoces',
    18: 'Books/Musics/Video',
    19: 'Food/Bevarage:Primatily e-commerce and shipping',
    20: 'Food/Bevarage:Primatily restaurant/Cafe/Food truck',
}
df['category'] = df.category.cat.rename_categories(new_categorits)

df['amount'] = df['amount'].replace({0: 1000, 1: 2000, 2: 3000})
df['tariff'] = df['tariff'].replace({0: 'VETURE', 1: 'BUSINESS', 2: 'UNLIMITED'})


df['validity'] = df.validity.astype('category')
new_categorits_for_validaty = {
    0: '1 month',
    1: '3 month',
    2: '6 month',
    3: '12 month',
    4: 'No limit'
}
df['validity'] = df.validity.cat.rename_categories(new_categorits_for_validaty)
df['month'] = df['time'].dt.month
```
